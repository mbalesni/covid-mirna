```python
string = 'GAUUGCUGCAGUCAUAACAAGAGAAGUGGGUUUUGUCGUGCCUGGUUUGCCUGGCACGAUAUUACGCACAACUAAUGGUGACUUUUUGCAUUUC'

def mfei_calculation(precursor, mfe):
    g_and_c = (precursor.count('G')+precursor.count('C'))/len(precursor)
    return mfe/len(precursor)/g_and_c

mfei_calculation(string,-37.3)
```

```python
file = open('default_miRNAFold.txt')

new_file = []
i = 0
for line in file:
    new_file.append(line.split())
```

```python
line = 0
mfe = []
sequences = []
while line < len(new_file):
    try:
        if line%2 == 0:
            mfe.append(new_file[line][2])
            sequences.append(new_file[line+1])
            line +=8
    except:
        break
```

```python
import ast
mfe = [ast.literal_eval(i) for i in mfe]
```

```python
mfe = [i[0] for i in mfe]
sequences = [i[0] for i in sequences]
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(list(zip(mfe, sequences)), columns =['MFE', 'sequence'])
```

```python
df
```

```python
mfeis = [mfei_calculation(sequences[i], mfe[i]) for i in range(len(mfe))]
df['MFEIS'] = mfeis
```

```python
df.head()
```

```python
#cutoff based on MFEIS value, I took very soft threshold, in biorxiv paper 
#it was suggested to go even more extreme to less than -0.85,
#but I for some reason (I think, cutoff point of 0.85 wasn't well explained for me. 
#I saw reasoning that it makes sense for MFEI of miRNA be higher than other coding 
#and non-coding RNAs as mRNAs (0.66), tRNAs (0.64) and rRNAs (0.59). So it makes sense to have > 0.66 (in absolute terms), but 
#-0.85 seems arbitrary. Anyways, the next tool will filter the ones that are left too, so it is not a big problem.)
filtered = df[df['MFEIS'] < -0.65]
```
